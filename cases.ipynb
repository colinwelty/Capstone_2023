```python
import xarray as xr
import numpy as np
```

```python
ds = xr.open_dataset("/Users/robertfrost/Documents/boomsoon/2023_semester_2/capstone/data/capstone_2023.nc")
```

```python
harvey = ds.where(ds.tc_name == "Harvey-17", drop=True)
```

```python
harvey_e = harvey.where(harvey.flash_number > 0, drop=True)
harvey_n = harvey.where(harvey.flash_number == 0, drop=True)
```

```python
he_rand = np.random.choice(len(harvey_e.mesocyclone_track_points))
hn_rand = np.random.choice(len(harvey_n.mesocyclone_track_points))
```

```python
he_idx = harvey_e.mesocyclone_id[he_rand].values
```

```python
hn_idx = harvey_n.mesocyclone_id[hn_rand].values
```

```python
lightning_case = ds.where(ds.mesocyclone_id == he_idx, drop=True)
nonlightning_case = ds.where(ds.mesocyclone_id == hn_idx, drop=True)
```
